# Price ì´ìƒì¹˜ ì œê±°í›„ì˜ íŒŒì¼: cleaned_naver.csv

```python
# priceê°€ ìˆ«ìí˜•ì¸ì§€ í™•ì¸ (í˜¹ì‹œ ë¬¸ìì—´ì´ë©´ ë³€í™˜)
df["price"] = pd.to_numeric(df["price"], errors="coerce")

# IQR(Interquartile Range) ë°©ì‹
Q1 = df["price"].quantile(0.25)  # 1ì‚¬ë¶„ìœ„ìˆ˜
Q3 = df["price"].quantile(0.75)  # 3ì‚¬ë¶„ìœ„ìˆ˜
IQR = Q3 - Q1

# ì´ìƒì¹˜ ê¸°ì¤€ (Q1 - 1.5*IQR ~ Q3 + 1.5*IQR ë²”ìœ„)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ì´ìƒì¹˜ ì œê±°
df = df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]

# ì¸ë±ìŠ¤ ë¦¬ì…‹
df = df.reset_index(drop=True)

# ì €ì¥
df.to_csv("./cleaned_naver.csv", index=False)
```

# price log 변화를 쓰는이유

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
data = pd.read_csv('../../csv/cleaned_total.csv')

data.head()
```

```python
data['price'].describe()
```

```python
price_data = data['price']

log_price_data = data['log_price']

print(log_price_data.head(), price_data.head)
```

```python
plt.figure(figsize=((7,4)))
sns.histplot(price_data, bins=50, kde=True, color="skyblue")
plt.xlabel("가격")
plt.ylabel("개수")
plt.title("가격 log 변화 전")
plt.show()
```

```python
plt.figure(figsize=((7,4)))
sns.histplot(log_price_data, bins=50, kde=True, color="skyblue")
plt.xlabel("가격")
plt.ylabel("개수")
plt.title("가격 log 변화 후")
plt.show()
```
